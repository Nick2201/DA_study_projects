```python
stock_data = pd.read_csv(r'.\data.csv.zip',compression='zip',encoding = 'ISO-8859-1')
stock_data.columns = [
    elem.lower() 
    for elem in stock_data.columns]
stock_data.columns    
stock_data.shape[0] - stock_data.drop_duplicates().shape[0]
```

```python
retail = stock_data.drop_duplicates()
retail.head()
retail_df_C = retail[retail.invoiceno.str.startswith('C')]
retail_df = retail[~retail.invoiceno.str.startswith('C')]
```

```python
def agg_bycustomer(_dataframe):
    agg_data_frame= (_dataframe 
        .groupby('customerid') 
        .agg(count_=("invoiceno",'count')))
    return agg_data_frame
agg_df_germnay = agg_bycustomer(retail.query('country == "Germany"'))
```

```python
import numpy as np
more_80 = np.percentile(agg_df_germnay,80)
germany_top = agg_df_germnay.reset_index().query('count_ > @more_80')['customerid']
```

```python
top_retail_germany = retail.query('customerid in @germany_top')
retail.query('customerid in @germany_top') \
    .groupby('invoiceno').agg(_count=('invoiceno','count')) \
    .query('_count > 1').index
```

```python
top_retail_germany \
    .groupby('stockcode').agg(preffer = ('stockcode','count')) \
    .drop('POST')
```
